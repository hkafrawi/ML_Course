````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Pima_indian_diabetes.csv')
df.head(20)
````

````python
# @title Glucose

from matplotlib import pyplot as plt
df['Glucose'].plot(kind='hist', bins=20, title='Glucose')
plt.gca().spines[['top', 'right',]].set_visible(False)
````

````python
print(df.shape)
df.describe()
````

````python
correlations = df.corr(method='pearson')
print(correlations)
````

````python
print(df.skew())
````

#Correlation Matrix

````python
f,ax = plt.subplots(figsize=[20,15])
sns.heatmap(df.corr(), annot=True, fmt = '.2f', ax=ax, cmap='viridis')
ax.set_title("Correlation Matrix", fontsize=20)
plt.show()
````

#  Histograms and density plots
*to add: density plot over histrograms, include skewness value*




````python
fig, ax = plt.subplots(4,2, figsize=(20,20)) #4 rows and 2 columns each
#Plotting histograms with overlaid density functions
sns.histplot(df.Pregnancies, bins=20, ax=ax[0,0], kde =True, color="blue")
sns.histplot(df.Glucose, bins=20, ax=ax[0,1], kde =True, color="blue")
sns.histplot(df.BloodPressure, bins=20, ax=ax[1,0], kde =True, color="blue")
sns.histplot(df.SkinThickness, bins=20, ax=ax[1,1], kde =True, color="blue")
sns.histplot(df.Insulin, bins=20, ax=ax[2,0], kde =True, color="blue")
sns.histplot(df.BMI, bins=20, ax=ax[2,1], kde =True, color="blue")
sns.histplot(df.DiabetesPedigreeFunction, bins=20, ax=ax[3,0], kde =True, color="blue")
sns.histplot(df.Age, bins=20, ax=ax[3,1], kde =True, color="blue")

# plotting skewness and kurtosis values in individual plots, by looping through each axes
#Source: https://stackoverflow.com/questions/50940283/show-metrics-like-kurtosis-skewness-on-distribution-plot-using-seaborn-in-pytho
for i, ax in enumerate(ax.reshape(-1)):
    ax.text(x=0.97, y=0.97, transform=ax.transAxes, s="Skewness: %f" % df.iloc[:,i].skew(),\
        fontweight='demibold', fontsize=15, verticalalignment='top', horizontalalignment='right',\
        backgroundcolor='white', color='red')
    ax.text(x=0.97, y=0.91, transform=ax.transAxes, s="Kurtosis: %f" % df.iloc[:,i].kurt(),\
        fontweight='demibold', fontsize=15, verticalalignment='top', horizontalalignment='right',\
        backgroundcolor='white', color='purple')
plt.tight_layout()
````
